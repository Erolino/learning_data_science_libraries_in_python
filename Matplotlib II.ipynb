```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

###### Functional Method

```python
#Functional Method
plt.plot(x,y)
#plt.show()
```

```python
plt.plot(x,y, 'r.') #additional arguments: color, style - .
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
#Multi plot
plt.subplot(1,2,1) # (number_of_rows, number_of_col, plot_number)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,y,'b')
```

###### Object Oriented API Method - better way

```python
#Object Oriented  - better way
fig = plt.figure() #imagenery set of canvas, we can as axes to this canvas
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) # left, bottom, width, height

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER')

axes2.plot(x,y)
axes2.set_title('smaller')
```

###### summary

### figsize, aspect ratio, DPI

```python
fig = plt.figure(figsize=(8,2),dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2),dpi=100)

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

###### legend

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,label='xy')
ax.plot(y,x,label='yx')

ax.legend(loc='best')
```

### Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,label='xy',color='red',lw=1,alpha=1,ls='steps')
ax.plot(y,x,label='yx',color='#FF8C66',
        linewidth=3,alpha=0.5,linestyle='-.',
        marker='o',markersize=8,markerfacecolor='purple')

ax.legend(loc='best')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,label='xy',color='red',lw=1,alpha=1,ls='-')
ax.plot(y,x,label='yx',color='#FF8C66',
        linewidth=3,alpha=0.5,linestyle='-.',
        marker='o',markersize=8,markerfacecolor='purple')

ax.legend(loc='best')
ax.set_xlim([0,2])
ax.set_ylim([0,2])
```
